<h3>K-Means Clustering Algorithm</h3>

<h3>Predict the optimum number of clusters</h3>

<h4>Preparing the Data</h4>

<h4>Create X</h4>

```python
x = iris_df.iloc[:, [0, 1, 2, 3,]].values
x
```

<h4>Create Y</h4>

<h3><font color=6295DC>Elbow Method</font></h3>

```python
#For creating y we need no of clusters. And we can find the optimum number of clusters using elbow method
```

<h4><font color=656966>-> In the Elbow method, we are actually varying the number of clusters ( K ) from 1 â€“ 10.<br/>-> For each value of K, we are calculating WCSS ( Within-Cluster Sum of Square ).<br/>-> WCSS is the sum of squared distance between each point and the centroid in a cluster.<br/>-> When we plot the WCSS with the K value, the plot looks like an Elbow.<br/>-> When we analyze the graph we can see that the graph will rapidly change at a point and thus creating an elbow shape.The K value corresponding to this point is the optimal K value or an optimal number of clusters</font></h4>

```python
within_cluster_sum_of_squares = []
Range_of_cluster = range(1, 11)
for i in Range_of_cluster:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x)
    within_cluster_sum_of_squares.append(kmeans.inertia_)
```

```python
plt.plot(Range_of_cluster, within_cluster_sum_of_squares,color="green")
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Within cluster sum of squares') 
plt.show()
```

<h4><font color=6295DC>The point at which the elbow shape is created is 3.Therefore optimal number of clusters is 3</font></h4>

<h4>K-Means Clustering</h4>

```python
y = KMeans(n_clusters=3)
```

<h4>Fit K means cluster model</h4>

```python
y_kmean = y.fit_predict(x)
y_kmean
```

<h3>Visualization of Results</h3>

<h4>Compare our original data vs clustered results</h4>

```python
new_labels = y.labels_
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(x[:, 0], x[:, 1], c=y_kmean, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[1].scatter(x[:, 0], x[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```
